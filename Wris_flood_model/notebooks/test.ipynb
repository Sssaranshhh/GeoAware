```python
import joblib
import pickle

# Path to existing model
old_model_path = "models/flood_model_stacking_ensemble.pkl"
new_model_path = "models/flood_model_stacking_ensemble.pkl"

# Load with pickle
with open(old_model_path, 'rb') as f:
    model_data = pickle.load(f)

# Re-save with joblib compression
joblib.dump(model_data, new_model_path, compress=9)

print("✅ Model compressed successfully!")
print(f"Saved to: {new_model_path}")
```

```python
import os

old_size = os.path.getsize("models/flood_model_stacking_ensemble.pkl")
print(f"File size: {old_size / 1024**2:.1f} MB")
```
